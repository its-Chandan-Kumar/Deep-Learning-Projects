# Import Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
import warnings
warnings.filterwarnings('ignore')
```

# Data Collection & Processing

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
df = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
df.head()
```

```python
df['label'] = breast_cancer_dataset.target
```

# Split Data

```python
# Split the dataset into input features and output labels
X = df.drop(columns='label',axis=1)
y = df['label']
```

```python
X.head()
```

```python
y.head()
```

```python
# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2,random_state=2)
```

```python
# Print the shapes of training and testing datasets
print(
    f'X_train: {X_train.shape}\n'
    f'X_test:  {X_test.shape}\n'
    f'y_train: {y_train.shape}\n'
    f'y_test:  {y_test.shape}'
)
```

```python
# Standardized the Data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

# Building Neural Network

```python
# importing Tensorflow & Keras

import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# Setting up the layers of my Neural Network

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# Compiling the Neural Network
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
# Training the Neural Network
history = model.fit(X_train_std,y_train,validation_split=0.1,epochs=15)
```

# Visualizing accuracy and loss

```python
model.summary()
```

```python
# Plot training and validation accuracy over epochs
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.legend(['training data','validation data'])
```

```python
# Plot training and validation loss over epochs
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epochs')

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['training data','validation data'])
```

```python
# Accuracy of the Model on test data
loss, accuracy = model.evaluate(X_test_std,y_test)

print(accuracy)
print(loss)
```

```python
# Print the shape of standardized test data and the first sample
print(X_test_std.shape)
print(X_test_std[0])
```

```python
# Generate predictions for the test dataset
y_pred = model.predict(X_test_std)
print(y_pred.shape)
print(y_pred[0])
```

```python
print(y_pred)
```

```python
# Converting the prediction probability to class labels
y_pred_labels = [np.argmax(i) for i in y_pred]
print(y_pred_labels)
```
